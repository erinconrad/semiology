```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.cluster import KMeans
from sklearn.decomposition import TruncatedSVD
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.utils import class_weight
from sklearn.metrics import roc_curve, auc
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegressionCV
from transformers import AutoTokenizer, AutoModel
tokenizer = AutoTokenizer.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")
model = AutoModel.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")




from sklearn.metrics import silhouette_score, davies_bouldin_score
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
import re
```

```python
# Load the CSV file into a DataFrame
df1 = pd.read_csv('../data/erin_semiology.csv')
```

```python
# Handling missing values
df1['semiology'] = df1['semiology'].fillna('')
```

```python
# Group by patient_id and aggregate semiology
df2 = df1.groupby('patient_id')['semiology'].agg(lambda x: ' '.join(set(x))).reset_index()
```

```python
# Merge df2 with df1 to retain original columns
df2 = pd.merge(df2, df1.drop_duplicates('patient_id'), on='patient_id', how='left')
```

```python
# Do some semiology preprocessing
def preprocess_text(text):
    
    # Convert text to lowercase
    text = text.lower()
    
    # remove a special thing
    text = re.sub(r'&gt;',' ',text)
    
    # remove semicolons, ampersands
    text = re.sub(r'[&;-]', ' ', text)
    
    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', ' ', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # other weird stuff to remove
    text = re.sub(r'\xa0',' ',text)
    
    # other weird stuff to remove
    text = re.sub(r'including aura',' ',text)
    
    start_index = text.find('description of emu evaluation')
    # If the phrase is found, remove it along with the subsequent characters
    if start_index != -1:
        text = text[:start_index]
    
    return text
```

```python
# Extract just the semiology bits from the templates
def extract_only_semiology(text):
    
    # flexible here
    text = re.sub('subjective semiology (including aura)','subjective semiology',text)
    
    sub_pattern = re.compile(r"(?<=subjective semiology)(.*?)(?=witnesses report)", re.DOTALL)
    sub_matches = sub_pattern.findall(text)
    
    wit_pattern = re.compile(r"(?<=witnesses report)(.*?)(?=duration)", re.DOTALL)
    wit_matches = wit_pattern.findall(text)
    
    if wit_matches or sub_matches:
        concatenated_text = ' '.join(sub_matches) + ' '.join(wit_matches)
        return concatenated_text
    else:
        return text
    
```

```python
# Other template
def extract_only_semiology2(text):
    
    
    sub_pattern = re.compile(r"(?<=event description)(.*?)(?=age at onset)", re.DOTALL)
    sub_matches = sub_pattern.findall(text)
    
    wit_pattern = re.compile(r"(?<=seizure description)(.*?)(?=age at onset)", re.DOTALL)
    wit_matches = wit_pattern.findall(text)
    
    if wit_matches or sub_matches:
        concatenated_text = ' '.join(sub_matches) + ' '.join(wit_matches)
        return concatenated_text
    else:
        return text
```

```python
def remove_extra_spaces(text):
    # Split the text into words
    words = text.split()
    # Join the words back together with a single space between them
    cleaned_text = ' '.join(words)
    # Remove any remaining extra spaces and newlines
    cleaned_text = cleaned_text.replace('\n', ' ')
    return cleaned_text
```

```python
# apply all cleaning
df2['clean_semiology'] = df2['semiology_x'].apply(preprocess_text)
df2['clean_semiology'] = df2['clean_semiology'].apply(extract_only_semiology)
df2['clean_semiology'] = df2['clean_semiology'].apply(extract_only_semiology2)
df2['clean_semiology'] = df2['clean_semiology'].apply(remove_extra_spaces)
```

```python
df3 = df2[df2['clean_semiology'] != '']
```

```python
# TF-IDF Vectorization for clustering
vectorizer = TfidfVectorizer(max_features=1000)  # You can adjust max_features as needed
tfidf_features = vectorizer.fit_transform(df3['clean_semiology'])

# K-means clustering
num_clusters = 5  # Adjust as needed
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(tfidf_features)


# Reduce dimensionality for visualization (optional)
svd = TruncatedSVD(n_components=2)
tfidf_features_reduced = svd.fit_transform(tfidf_features)

# Plot clusters
plt.figure(figsize=(8, 6))
for cluster_label in np.unique(cluster_labels):
    cluster_indices = np.where(cluster_labels == cluster_label)[0]
    plt.scatter(tfidf_features_reduced[cluster_indices, 0],
                tfidf_features_reduced[cluster_indices, 1],
                label=f'Cluster {cluster_label}',
                alpha=0.7)

# Plot cluster means (centroids)
cluster_centers = kmeans.cluster_centers_
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1],
            marker='x', s=100, c='k', label='Cluster Centers')

# Annotate label descriptions for cluster 1
cluster_1_indices = np.where(cluster_labels == 1)[0]

plt.title('K-means Clustering')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Restrict to just epilepsy or pnes
df4 = df3[df3['single_diagnosis'].isin(['epilepsy', 'pnes'])]
```

```python
# Make a new censored_semiology which censors obvious seizure words
def censor_text(text):
    
    phrases_to_replace = ['fias','seizures','events','event','seizure','btc','btcs','fbtcs','aura','gtc']
    
    # Construct a regular expression pattern by joining the phrases with "|" to match any of them
    pattern = '|'.join(re.escape(phrase) for phrase in phrases_to_replace)
    
    # Use the sub() function to replace matched phrases with an empty string
    cleaned_text = re.sub(pattern, '', text)
    
    return cleaned_text
```

```python
df4['censored_semiology'] = df4['clean_semiology'].apply(censor_text)
#df4['censored_semiology'] = df4['clean_semiology']
```

```python
y = df4['single_diagnosis']
```

```python
class_counts = df4['single_diagnosis'].value_counts()
print(class_counts)
```

```python
# Machine learning to predict epilepsy vs pnee

X = vectorizer.fit_transform(df4['censored_semiology']).toarray()
y = df4['single_diagnosis']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)

# Map class labels to binary labels
y_train_binary = y_train.map({'epilepsy': 0, 'pnes': 1})
y_test_binary = y_test.map({'epilepsy': 0, 'pnes': 1})

# Define class weights
class_weights = {0: 1.0, 1: len(y_train) / (2 * sum(y_train == 'pnes'))}

# Create and train a logistic regression classifier with class weights
#log_reg = LogisticRegressionCV(penalty='l1', solver='liblinear', cv=5,class_weight=class_weights)
log_reg = LogisticRegression(class_weight=class_weights)
log_reg.fit(X_train, y_train_binary)

# Calculate predicted probabilities
y_probs = log_reg.predict_proba(X_test)[:, 1]

# Calculate the ROC curve
fpr, tpr, thresholds = roc_curve(y_test_binary, y_probs)

# Calculate the area under the ROC curve
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
# Predict classes
y_pred = log_reg.predict(X_test)

# Create confusion matrix
conf_matrix = confusion_matrix(y_test_binary, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", cbar=False, 
            xticklabels=['epilepsy', 'pnes'], yticklabels=['epilepsy', 'pnes'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# see highest prob cases of pnes
# Create a DataFrame to store predicted probabilities and original labels
predictions_df = pd.DataFrame({'Probability_pnes': y_probs, 'Probability_epilepsy':1-y_probs,'Actual_Label': y_test})

# Sort instances based on predicted probabilities (descending order)
predictions_df_sorted = predictions_df.sort_values(by='Probability_pnes', ascending=False)
predictions_df_reverse_sorted = predictions_df.sort_values(by='Probability_epilepsy', ascending=False)


# Display instances with the highest probability of being "pnes"
top_n = 10  # You can adjust this value to display more or fewer instances
top_pnes_instances = predictions_df_sorted.head(top_n)
top_epilepsy_instances = predictions_df_reverse_sorted.head(top_n)
print("Cases with the highest probability of being pnes:")
print(top_pnes_instances)

print("Cases with the highest probability of being epilepsy:")
print(top_epilepsy_instances)
```

```python
# Merge the DataFrame containing predicted probabilities with the original DataFrame
top_pnes_instances_with_text = pd.merge(top_pnes_instances, df4, left_index=True, right_index=True)
top_epilepsy_instances_with_text = pd.merge(top_epilepsy_instances, df4, left_index=True, right_index=True)


# Select relevant columns (e.g., "cleaner_semiology", "Probability_pnes", "Actual_Label")
relevant_columns = ["clean_semiology","censored_semiology", "Probability_pnes", "Probability_epilepsy", "Actual_Label"]
top_pnes_instances_text = top_pnes_instances_with_text[relevant_columns]
top_epilepsy_instances_text = top_epilepsy_instances_with_text[relevant_columns]


# Display instances with the highest probability of being "pnes" along with their text
print("Cases with the highest probability of being pnes and their corresponding text:")
print(top_pnes_instances_text)

print("Cases with the highest probability of being epilepsy and their corresponding text:")
print(top_epilepsy_instances_text)
```
