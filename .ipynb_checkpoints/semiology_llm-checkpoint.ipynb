```python
import numpy as np
import pandas as pd
import re
from torch.utils.data import DataLoader, Dataset
import torch
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer, AutoModel, Trainer, TrainingArguments, AutoModelForSequenceClassification, AutoConfig, BertTokenizer, BertForSequenceClassification
from torch.nn.functional import softmax
from torch.utils.data import DataLoader
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
```

```python
# Define preprocessing functions

# General preprocessing
def preprocess_text(text):
    
    # Convert text to lowercase
    text = text.lower()
    
    # remove a special thing
    text = re.sub(r'&gt;',' ',text)
    
    # remove semicolons, ampersands
    text = re.sub(r'[&;-]', ' ', text)
    
    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', ' ', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # other weird stuff to remove
    text = re.sub(r'\xa0',' ',text)
    
    # other weird stuff to remove
    text = re.sub(r'including aura',' ',text)
    
    start_index = text.find('description of emu evaluation')
    # If the phrase is found, remove it along with the subsequent characters
    if start_index != -1:
        text = text[:start_index]
    
    return text

# Extract just the semiology bits from the templates
def extract_only_semiology(text):
    
    # flexible here
    text = re.sub('subjective semiology (including aura)','subjective semiology',text)
    
    sub_pattern = re.compile(r"(?<=subjective semiology)(.*?)(?=witnesses report)", re.DOTALL)
    sub_matches = sub_pattern.findall(text)
    
    wit_pattern = re.compile(r"(?<=witnesses report)(.*?)(?=duration)", re.DOTALL)
    wit_matches = wit_pattern.findall(text)
    
    if wit_matches or sub_matches:
        concatenated_text = ' '.join(sub_matches) + ' '.join(wit_matches)
        return concatenated_text
    else:
        return text
    
# Other template
def extract_only_semiology2(text):
    
    
    sub_pattern = re.compile(r"(?<=event description)(.*?)(?=age at onset)", re.DOTALL)
    sub_matches = sub_pattern.findall(text)
    
    wit_pattern = re.compile(r"(?<=seizure description)(.*?)(?=age at onset)", re.DOTALL)
    wit_matches = wit_pattern.findall(text)
    
    if wit_matches or sub_matches:
        concatenated_text = ' '.join(sub_matches) + ' '.join(wit_matches)
        return concatenated_text
    else:
        return text
    
def remove_extra_spaces(text):
    # Split the text into words
    words = text.split()
    # Join the words back together with a single space between them
    cleaned_text = ' '.join(words)
    # Remove any remaining extra spaces and newlines
    cleaned_text = cleaned_text.replace('\n', ' ')
    return cleaned_text

# Make a new censored_semiology which censors obvious seizure words
def censor_text(text):
    
    phrases_to_replace = ['fias','seizures','events','event','seizure','btc','btcs','fbtcs','aura','gtc']
    
    # Construct a regular expression pattern by joining the phrases with "|" to match any of them
    pattern = '|'.join(re.escape(phrase) for phrase in phrases_to_replace)
    
    # Use the sub() function to replace matched phrases with an empty string
    cleaned_text = re.sub(pattern, '', text)
    
    return cleaned_text
```

```python
# Load in and pre-process data

# Load the CSV file into a DataFrame
df1 = pd.read_csv('../data/erin_semiology.csv')

# Handling missing values
df1['semiology'] = df1['semiology'].fillna('')

# Group by patient_id and aggregate semiology
df2 = df1.groupby('patient_id')['semiology'].agg(lambda x: ' '.join(set(x))).reset_index()

# Merge df2 with df1 to retain original columns
df2 = pd.merge(df2, df1.drop_duplicates('patient_id'), on='patient_id', how='left')

# apply all cleaning
df2['clean_semiology'] = df2['semiology_x'].apply(preprocess_text)
df2['clean_semiology'] = df2['clean_semiology'].apply(extract_only_semiology)
df2['clean_semiology'] = df2['clean_semiology'].apply(extract_only_semiology2)
df2['clean_semiology'] = df2['clean_semiology'].apply(remove_extra_spaces)

df3 = df2[df2['clean_semiology'] != '']

# Restrict to just epilepsy or pnes
df4 = df3[df3['single_diagnosis'].isin(['epilepsy', 'pnes'])]

# Censor text
df4['censored_semiology'] = df4['clean_semiology'].apply(censor_text)

# Turn labels into ones and zeros
df4['labels'] = df4['single_diagnosis'].apply(lambda x: 1 if x == 'epilepsy' else 0)

# Define final df
df = df4
```

```python
# Define your dataset class
class TextDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_length=512):
        self.tokenizer = tokenizer
        self.data = dataframe
        self.texts = dataframe['censored_semiology'].tolist()
        self.labels = dataframe['labels'].tolist()
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        labels = self.labels[idx]
        encoding = self.tokenizer(text, padding='max_length', truncation=True, max_length=self.max_length, return_tensors="pt")
        
        # Here we assume 'labels' is already numeric. If not, you'll need to convert it.
        return {
            'input_ids': encoding['input_ids'].flatten(),  # Flatten to remove the batch dimension
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(labels, dtype=torch.long)
        }
```

```python
# Split data into training and testing
train_df, test_df = train_test_split(df, test_size=0.4, random_state=42)  # 60% training, 40% testing
```

```python
# Import BioClinical BERT tokenizer
tokenizer = AutoTokenizer.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")

# Load the configuration of Bio_ClinicalBERT
config = AutoConfig.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")

# Assuming you have 2 labels for your classification task
config.num_labels = 2

# Load Bio_ClinicalBERT with a sequence classification head
model = AutoModelForSequenceClassification.from_pretrained("emilyalsentzer/Bio_ClinicalBERT", config=config)
```

```python
# Tokenize and prepare the data
train_dataset = TextDataset(train_df, tokenizer)
```

```python
# Do training

# Define the training arguments
training_args = TrainingArguments(
    output_dir='../results/bioclinical_bert/',          # output directory
    num_train_epochs=3,              # total number of training epochs
    per_device_train_batch_size=8,  # batch size per device during training
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./logs',            # directory for storing logs
    logging_steps=10,
)

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
)

# Train the model
trainer.train()

# Save the fine-tuned model
model.save_pretrained('../results/bioclinical_bert/fine_tuned_model')
```

```python
# Save the tokenizer
tokenizer.save_pretrained('../results/bioclinical_bert/fine_tuned_model')
```

```python
# Load the tokenizer and fine tuned model
tokenizer = BertTokenizer.from_pretrained('../results/bioclinical_bert/fine_tuned_model')
model = BertForSequenceClassification.from_pretrained('../results/bioclinical_bert/fine_tuned_model')                                        
```

```python
# Tokenize and prepare the data
test_dataset = TextDataset(test_df, tokenizer)
```

```python
test_dataloader = DataLoader(test_dataset, batch_size=16)  # Adjust batch_size according to your needs
```

```python
# Test model on test data
model.eval()

predictions = []
labels = []

with torch.no_grad():
    for batch in test_dataloader:
        inputs = {k: v.to(model.device) for k, v in batch.items() if k != 'labels'}
        output = model(**inputs)
        logits = output.logits
        probabilities = torch.nn.functional.softmax(logits, dim=-1)
        predictions.extend(probabilities.cpu().numpy())
        labels.extend(batch['labels'].cpu().numpy())
```

```python
# Make ROC curve
y_true = labels
y_score = [p[1] for p in predictions]  # Probability of the positive class

fpr, tpr, thresholds = roc_curve(y_true, y_score)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
